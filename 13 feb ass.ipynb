```python
''' error defined in arithmrticerror are
OverFlowError
ZeroDivisionError
FloatingPointError'''

#OverFlowError
try:
    print(5/(10.2**10000000000))
except OverflowError  as e:
    print(e)

#ZeroDivisionError
try:
    5/0
except ZeroDivisionError as e:
    print(e)
```

```python
'''
The LookupError exception in Python forms the base class for all exceptions that are raised 
when an index or a key is not found for a sequence or dictionary respectively.'''

l=[1,2,3,4]
d={'key1':"value1",'key2':"value2",'key3':"value3"}
try:
    print(l[4])
except  LookupError as e:
    print(e)
try:
    d['key4']
except  LookupError as e:
    print(e)
    
```
